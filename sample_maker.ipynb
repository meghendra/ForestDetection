```python
# 100 X 100 samples from large tif
import os
from skimage.external.tifffile import imread, imshow
import scipy.misc
import image_slicer

path = ".\\more_samples\\f\\"
listing = os.listdir(path)
for file in listing:
    im = imread(path+file)
    x,y,z = im.shape
    x_crop = (x // 100) * 100
    y_crop = (y // 100) * 100
    m=min([x_crop,y_crop])
    print(file)
    print(x_crop,y_crop,m)
    t = (m * m // 10000)
    print(t)
    crop_im = im[0:x_crop, 0:y_crop]
    scipy.misc.toimage(crop_im, cmin=0.0, cmax=...).save(path+file+".jpg")
    img = path+file+".jpg"
    tiles = image_slicer.slice(img,t,save=False)
    image_slicer.save_tiles(tiles, directory='.\\more_samples\\forest_slices_100')
```

```python
# 10 X 10 samples from 100 X 100 samples
import os
from PIL import Image
import scipy.misc
import image_slicer

path = ".\\more_data\\forests\\"
listing = os.listdir(path)
i=0
for file in listing:
    im = Image.open(path+file)
    x = im.size[0]
    y = im.size[1]
    x_crop = (x // 10) * 10
    y_crop = (y // 10) * 10
    m=min([x_crop,y_crop])
    print(file)
    print(x_crop,y_crop,m)
    t = (m * m // 100)
    print(t)
    crop_im = im.crop((0,0,100,100))
    scipy.misc.toimage(crop_im, cmin=0.0, cmax=...).save(".\\more_data\\jpegs\\forests\\"+file+".jpg")
    img = ".\\more_data\\jpegs\\forests\\"+file+".jpg"
    tiles = image_slicer.slice(img,100,save=False)
    image_slicer.save_tiles(tiles, directory='.\\more_data\\10\\forests',prefix=str(i)+"_")
    i+=1
```
